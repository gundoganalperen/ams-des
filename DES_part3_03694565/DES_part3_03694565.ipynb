### Analysis and General Questions

#### Task 3.3.1: Seed for Random Number Generation

* Using the same seed in the simulation guarantees the same trace of the random variable during a simulation run which will give the same simulation results that can be repeated. Then, we can compare the different simulations with the same seed. 

#### Task 3.3.2: Bins for Histograms

* If I increase the number of bins, then there will be many peaks in the simulation which makes it harder to observe distribution. If bins are small, then all the data will gather in same spot which will make it difficult to observe the behaviour of the data. There are different rules (e.g. the Freedman-Diaconis Rule, Sturges' Rule, Scott's Rule, etc.) explaining how to choose bins. However, bins should be depend on the number of samples e.g. if the samples are high, then it is better to have more bins to observe the distribution. 

#### Task 3.3.2: System Utilization
