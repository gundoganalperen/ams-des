## 4.4 Analysis and General Questions

### Task 4.4.1: Lag size

* First sequence repeats itself on every 2 element and provides no new information after that, therefore 0,1 lag size will be enough.
* For the second sequence, 0,1,2 lag sizes will be enough to observe the behaviour of the data, because it repeat itself in every 3 element

### Task 4.4.2: Correlation Times in the DES

* **Correlation between IAT and waiting time of a packet:** Correlation between them is negative and not so high. So, if a packet arrives within short intervals, then it has to wait more.
* **Correlation between IAT and serving time of a packet:** It is very small and it can be considered as zero. This is actually expected result because both IAT and serving units are two different random processes so that there should be no correlation between them.
* **Correlation between IAT and system time (waiting time + serving time) of a packet:** System time is consists of waiting time and serving time, there is no correlation between IAT and serving time as I explained above but there is correlation between IAT and WT. Therefore, this will perform also negative correlation between IAT and system time. So, if IAT intervals are short then system time will be higher.
* **Correlation between serving time and system time of a packet:** As printed above, the correlation between ST and Syst. is high for low rho values and low for higher rho values. 
* **Auto-correlation of waiting time with lags ranging from 1 to 20:** The auto-correlation of WT is depend on the RHO value, for larger rho values correlation is really high. For higher rho values, arrival rate will be higher so that packets most likely have to wait more time. 

### Task 4.4.3: Visualization of Covariance/Correlation

* If there is no correlation, then the points in the scatter plot will be independent from each other. If there is a positive correlation(e.g. 0.99) then the points will be on the same line but skewed on the right. Also, If there is a negative correlation, than the shape of the graph skewed on the left. So, correlation affects the behaviour of dots in the scatter plots.
